```
# 2d Cellular Automata 
# Visualization of 3 generations on a square grid
# New generation is a 'Convolude' with a rules signal 
init_livecells_percent = 0.05 
rule = np.array([[1,0,1],[1,0,1],[1,0,0]])
gridD = 50
marker_size = 20
from scipy.signal import convolve2d

import warnings
warnings. filterwarnings("ignore") 

import matplotlib as matlib
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(5, 5)) # contains total four grids 
ax0 = fig.add_subplot(221); ax0.set_title("seed")
ax1 = fig.add_subplot(222); ax1.set_title("gen_1")
ax2 = fig.add_subplot(223); ax2.set_title("gen_2")
ax3 = fig.add_subplot(224); ax3.set_title("gen_3")
ax0.get_xaxis().set_visible(False); ax0.get_yaxis().set_visible(False)
ax1.get_xaxis().set_visible(False); ax1.get_yaxis().set_visible(False)
ax2.get_xaxis().set_visible(False); ax2.get_yaxis().set_visible(False)
ax3.get_xaxis().set_visible(False); ax3.get_yaxis().set_visible(False)
scatter0 = ax0.scatter(*board, animated=True, s=marker_size, edgecolor="silver")
scatter1 = ax1.scatter(*board, animated=True, s=marker_size, edgecolor="white")
scatter2 = ax2.scatter(*board, animated=True, s=marker_size, edgecolor="white")
scatter3 = ax3.scatter(*board, animated=True, s=marker_size, edgecolor="white")

import numpy as np
seed = np.random.uniform(0, 1, size=(gridD,gridD)) <= init_livecells_percent

# Game of Life as a scatter plot on the 2d board
xs = np.arange(0, gridD); ys = np.arange(0, gridD)
board = np.meshgrid(xs, ys)
colormap=[matlib.cm.spring(N) for N in (seed).flatten()]
color = [("gold") if z1 else z2 for z1, z2 in zip((seed).flatten(),colormap)]
scatter0.set_facecolor(color)

gen1 = convolve2d(seed, rule, mode='same', boundary="wrap")

# Board after running generation 1 
color = [("gold") if z1 else z2 for z1, z2 in zip((gen1).flatten(), colormap)]
scatter1.set_facecolor(color)

# Board after running generation 2 
gen2 = convolve2d(gen1, rule, mode='same', boundary="wrap")
color = [("green") if z1 else z2 for z1, z2 in zip((gen2).flatten(), colormap)]
scatter2.set_facecolor(color)

# Board after running generation 3 
gen3 = convolve2d(gen2, rule, mode='same', boundary="wrap")
color = [("white") if z1 else z2 for z1, z2 in zip((gen3).flatten(), colormap)]
scatter3.set_facecolor(color)

```
